# Pyplot Example

#### Import Functions from Numpy and PyPlot

```python
from numpy import arange, meshgrid, pi, sin, cos
from matplotlib.pyplot import figure, contour, contourf
```

#### Show Images inside the Notebook

```python
%matplotlib inline
```

#### Create Arrays

```python
x = arange(-4 * pi, 4 * pi, 0.1)
y = arange(-4 * pi, 4 * pi, 0.1)
```

#### Create Matrices

```python
X, Y = meshgrid(x, y)
```

#### Apply Function

```python
Z = cos(X) + cos(Y)
```

#### Create Figure

```python
fig = figure(figsize=(5, 5))
contour(X, Y, Z)
```

```python
fig = figure(figsize=(5,5))
contourf(X, Y, Z)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
fig = figure(figsize=(10, 5))
p3d = fig.add_subplot(111, projection='3d')
```

```python
colormap = cm.Spectral_r
p3d.plot_surface(X, Y, Z, cstride=3, rstride=3, cmap=colormap)
```
