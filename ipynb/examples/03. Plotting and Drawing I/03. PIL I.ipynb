```python
# import stuff
import PIL
from PIL import Image
from botsnplots import loadImage
```

```python
# download a file from the web
myFile = loadImage("https://upload.wikimedia.org/wikipedia/en/c/cb/Marvin_%28HHGG%29.jpg")
```

```python
# open file as image + show it
pic = Image.open(myFile)
pic
```

```python
# show image size
xsize, ysize = pic.size
xsize, ysize
```

```python
# load image
pic = Image.open(myFile)

# get pixels as array
pixels = pic.load()

# black white limit
limit = 200

for x in range(xsize):
    for y in range(ysize):
        r, g, b = pixels[x, y]
        if (r + g + b) / 3.0 < limit:
            pixels[x, y] = (0, 0, 0)
        else:
            pixels[x, y] = (255, 255, 255)
        
pic
```

```python
pic.save("marvin-black-white.png")
```
