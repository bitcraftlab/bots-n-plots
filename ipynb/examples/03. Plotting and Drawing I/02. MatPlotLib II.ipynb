# Pyplot Example

#### Use PyLab inside the Notebook

```python
%pylab inline
```

#### Create Arrays

```python
x = linspace(-4 * pi, 4 * pi, 200)
y = linspace(-4 * pi, 4 * pi, 200)
```

#### Create Matrices

```python
X, Y = meshgrid(x, y)
```

#### Apply Function

```python
Z = cos(X) + cos(Y)
```

#### Create Figure

```python
fig = figure(figsize=(5, 5))
contour(X, Y, Z)
```

```python
fig = figure(figsize=(5,5))
contourf(X, Y, Z)
```
